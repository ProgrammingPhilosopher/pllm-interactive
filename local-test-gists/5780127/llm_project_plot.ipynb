```python
# run this locally – needs numpy/matplotlib
import matplotlib.pyplot as plt

labels     = ['Original PLLM\n(200 fails)', 'Enhanced PLLM\n(200 same gists)']
successes  = [0, 76]
failures   = [200, 124]

fig, ax = plt.subplots(figsize=(5,4))
ax.bar(labels, successes, label='Success', color='tab:blue')
ax.bar(labels, failures, bottom=successes, label='Fail', color='tab:red')

# annotate % on top of bars
for x, s, f in zip(labels, successes, failures):
    pct = 100 * s / (s+f)
    ax.text(x, s+f+2, f'{pct:.1f} %', ha='center', fontweight='bold')

ax.set_ylabel('Number of gists (out of 200)')
ax.set_title('Enhanced PLLM rescues 38 % of originally‑failing gists')
ax.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
